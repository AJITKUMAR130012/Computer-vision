```python
import numpy as np
import pandas as pd
import cv2
import easyocr
import matplotlib.pyplot as plt
```

```python
img1=cv2.imread("./Dataset/Screenshot 2023-01-08 at 5.14.24 PM.png")
img2=cv2.imread("./Dataset/Screenshot 2023-01-08 at 5.14.47 PM.png")
```

```python
plt.imshow(img1)
plt.show()
plt.imshow(img2)
plt.show()
```

```python
reader = easyocr.Reader(['en'])
result1 = reader.readtext(img1)
result2= reader.readtext(img2)
```
