## Difference Method

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load the reference and test images
ref_image = cv2.imread("Dataset/Problem-3/example-1/reference.png")
test_image = cv2.imread("Dataset/Problem-3/example-1/perfect/perfect1.png")

#resize the image

ref_image = cv2.resize(ref_image,(256,256))
test_image =cv2.resize(test_image,(256,256))
plt.imshow(test_image)
plt.show()

# Convert the images to grayscale
gray_ref_image = cv2.cvtColor(ref_image, cv2.COLOR_BGR2GRAY)
gray_test_image = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)


# Compute the absolute difference between the two grayscale images
diff_image = cv2.absdiff(gray_ref_image, gray_test_image)

# Apply adaptive thresholding to the difference image to create a binary image
binary_image = cv2.adaptiveThreshold(diff_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 35)

# Apply morphological operations to remove noise and fill gaps in the binary image
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
# print(kernel)
morph_image = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel)
# plt.imshow(morph_image)
# plt.show()
# Find the contours of the binary image
contours, _ = cv2.findContours(morph_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw a rectangle around each contour on the test image
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(test_image, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Display the test image with the faulty regions highlighted
plt.imshow( test_image)
plt.show()
```

## Image registration and Difference Method

```python
import cv2

# Load the two images to be registered
reference = cv2.imread('Dataset/Problem-3/example-1/reference.png')
perfect1 = cv2.imread('Dataset/Problem-3/example-1/perfect/perfect1.png')

# Convert the images to grayscale
image1_gray = cv2.cvtColor(reference, cv2.COLOR_BGR2GRAY)
image2_gray = cv2.cvtColor(perfect1, cv2.COLOR_BGR2GRAY)

# Find the keypoints and descriptors with SIFT
sift = cv2.SIFT_create()
keypoints1, descriptors1 = sift.detectAndCompute(image1_gray, None)
keypoints2, descriptors2 = sift.detectAndCompute(image2_gray, None)

# Create a BFMatcher object and match the descriptors
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)
matches = bf.match(descriptors1, descriptors2)

# Sort the matches by distance
matches = sorted(matches, key=lambda x: x.distance)

# Extract the matched keypoints
points1 = []
points2 = []
for match in matches:
    points1.append(keypoints1[match.queryIdx].pt)
    points2.append(keypoints2[match.trainIdx].pt)

# Convert the points to numpy arrays
points1 = np.float32(points1)
points2 = np.float32(points2)

# Find the homography matrix
homography, _ = cv2.findHomography(points1, points2, cv2.RANSAC)

# Warp image 1 to image 2 using the homography matrix
registered_image = cv2.warpPerspective(perfect1, homography, (reference.shape[1], reference.shape[0]))

# Show the registered image
plt.imshow( registered_image)
plt.show()
plt.imshow(reference-registered_image)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the two images to be registered
image1 = cv2.imread('Dataset/Problem-3/example-1/reference.png')
image2 = cv2.imread('Dataset/Problem-3/example-1/perfect/perfect1.png')

# Convert the images to grayscale
image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Define the ratio of the two images
ratio = image1.shape[0] / image2.shape[0]

# Initialize the SIFT detector
sift = cv2.SIFT_create()

# Detect keypoints and compute descriptors for both images
keypoints1, descriptors1 = sift.detectAndCompute(image1_gray, None)
keypoints2, descriptors2 = sift.detectAndCompute(image2_gray, None)

# Match the descriptors using the Brute-Force matcher
bf = cv2.BFMatcher()
matches = bf.knnMatch(descriptors1, descriptors2, k=2)

# Apply the ratio test to select only good matches
good_matches = []
for m, n in matches:
    if m.distance < 0.75 * n.distance:
        good_matches.append(m)

# Extract the keypoints coordinates for the good matches
source_points = np.float32([keypoints1[m.queryIdx].pt for m in good_matches])
destination_points = np.float32([keypoints2[m.trainIdx].pt for m in good_matches])

# Scale the source points to match the size of the destination image
source_points_scaled = np.float32([[point[0] / ratio, point[1] / ratio] for point in source_points])

# Compute the affine transformation matrix
M, _ = cv2.estimateAffinePartial2D(source_points_scaled, destination_points)

homography, _ = cv2.findHomography(source_points_scaled, destination_points, cv2.RANSAC)

# Warp image 2 to image 1 using the affine transformation matrix
#registered_image = cv2.warpAffine(image2, M, (image1.shape[1], image1.shape[0]))

registered_image = cv2.warpPerspective(image2, homography, (image1.shape[1], image1.shape[0]))


# Show the registered image
plt.imshow( registered_image)
plt.show()
plt.imshow(image1-registered_image)
plt.show()
```
